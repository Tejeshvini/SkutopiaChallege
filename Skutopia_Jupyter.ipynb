```python
## Insert the data extracted from the API into the postgres database
def execute_values(conn, df, table): 
  
    tuples = [tuple(x) for x in df.to_numpy()] 
  
    cols = ','.join(list(df.columns)) 
    # SQL query to execute 
    query = "INSERT INTO %s(%s) VALUES %%s" % (table, cols) 
    cursor = conn.cursor() 
    try: 
        extras.execute_values(cursor, query, tuples) 
        conn.commit() 
    except (Exception, psycopg2.DatabaseError) as error: 
        print("Error: %s" % error) 
        conn.rollback() 
        cursor.close() 
        return 1
    print("the dataframe is inserted") 
    cursor.close() 
```

```python
cancels = online_retail[online_retail['InvoiceNo'].str.contains("C")]
```

```python
## definig a function to create aggregated tables for all sales (including returns) and returns only, 
## ranking the results by Revenue
def ranked(x):
    x['Revenue'] = x['Quantity'].abs()*x['UnitPrice']
    x_grouped = pd.DataFrame(x.groupby('Description')['Revenue'].sum()).reset_index()
    x_grouped['Rank']=x_grouped['Revenue'].rank(ascending=False)
    return(x_grouped.sort_values(by=['Rank']))
```

```python
cancels_grouped_ranked = ranked(cancels)
all_grouped_ranked = ranked(online_retail)
```
